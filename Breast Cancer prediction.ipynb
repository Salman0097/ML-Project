### Encoding the values in diagnosis column

- 1 --> Benign => Patient do not have cancer
- 0 --> Malignant => Patient has cancer

```python
breast_cancer['diagnosis'] = breast_cancer['diagnosis'].replace({'B': 1 ,'M':0})
```

## Features and Labels

```python
X = breast_cancer.iloc[ : , 1:]
y = breast_cancer['diagnosis']
```

```python
X.head(3)
```

```python
y.head(3)
```

## Splitting the training and testing datas

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, stratify=y)
```

```python
X.shape, X_train.shape,X_test.shape, y_train.shape, y_test.shape
```

## Training and fitting the data

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train, y_train)
```

## Predict the test data using the trained model 

```python
y_pred = model.predict(X_test)
print('The predicted values are : \n',y_pred)
```

## Checking the accuracy score between predicted and test labels

```python
from sklearn.metrics import accuracy_score
```

```python
acc_sc = accuracy_score(y_test, y_pred)
print(f'The accuracy score of predicted outcomes with the testing outcomes is : {acc_sc}')
```
